# The magical function to scrape info for each movie

```python
def get_movie_info(link):
    """
    Takes in a string of url of a movie, collect information including 'domestic title gross',
    'movie title', 'rating', 'release date', and 'runtime', and creates dictionary
    of information, then appends to movie_data list of dicts.
    """
    
    def get_movie_value(soup, field_name):
        """Grab a value from boxofficemojo HTML
        Takes a string attribute of a movie on the page and
        returns the string in the next sibling object
        (the value for that attribute)
        or None if nothing is found.
        """
        obj = soup.find(text=re.compile(field_name))
        if not obj:
            return(None)
        next_sibling = obj.findNextSibling()
        if next_sibling:
            return(next_sibling.text)
        else:
            return(None)
        
    def to_date(datestring):
        """Takes in a date in string form and returns it as a datetime type"""
        date = dateutil.parser.parse(datestring)
        return(date)
    
    def money_to_int(moneystring):
        """Takes in an amount of money in string form and returns it as a clean integer"""
        if moneystring == None:
            return(None)
        else:
            moneystring = moneystring.replace('$', '').replace(',', '')
            return(int(moneystring))

    def runtime_to_minutes(runtimestring):
        """Takes in the runtime in hours and minutes and returns runtime in minutes"""
        runtime = runtimestring.split()
        try:
            minutes = int(runtime[0])*60 + int(runtime[2])
            return(minutes)
        except:
            return(None)
        
    def convert_budget(bdgt):
        """Takes in a budget with the word million and returns it as a clean integer"""
        if bdgt == 'N/A':
            return(None)
        elif 'million' in bdgt:
            budget_string = bdgt.replace('$','')
            amount = (budget_string.split(' ')[1])
            try:
                budget_int = int(budget_string.split(' ')[0])
                budget_int = budget_int * 1000000
                return(budget_int)
            except ValueError:
                return(None)
        else:
            return(None)
    
    def get_subgenre(movie_tit):
        """Takes in a movie title and returns genres of that movie"""
        genres = []
        if movie_tit in animation_scifi_dict:
            genres.append('animated')
            genres.append('sci-fi')
        if movie_tit in adventure_dict:
            genres.append('adventure')
            genres.append('sci-fi')
        if movie_tit in alien_dict:
            genres.append('alien invasion')
            genres.append('sci-fi')
        if movie_tit in book_dict:
            genres.append('based on a book')
            genres.append('sci-fi')
        if movie_tit in satire_dict:
            genres.append('satire')
            genres.append('sci-fi')
        if movie_tit in chase_dict:
            genres.append('chase')
            genres.append('sci-fi')
        if movie_tit in comedy_dict:
            genres.append('comedy')
            genres.append('sci-fi')
        if movie_tit in horror_dict:
            genres.append('horror')
            genres.append('sci-fi')
        if movie_tit in remake_dict:
            genres.append('remake')
            genres.append('sci-fi')
        if movie_tit in threeD_dict:
            genres.append('3D')
        if movie_tit in christian_dict:
            genres.append('christian')
        if movie_tit in black_com_dict:
            genres.append('black comedy')
        if movie_tit in comic_dict:
            genres.append('adapted from a comic book')
        if movie_tit in crime_dict:
            genres.append('crime')
        if movie_tit in dance_dict:
            genres.append('dance')
        if movie_tit in documentary_dict:
            genres.append('documentary')
        if movie_tit in foreign_dict:
            genres.append('foreign film')
        if movie_tit in gay_dict:
            genres.append('gay/lesbian')
        if movie_tit in romcom_dict:
            genres.append('romantic comedy')
        if movie_tit in spy_dict:
            genres.append('spy')
        if movie_tit in sports_dict:
            genres.append('sports drama')
        if movie_tit in superhero_dict:
            genres.append('superhero')
        return(genres)
    
    response = requests.get(link)
    page = response.text
    soup = BeautifulSoup(page, 'lxml')
    
    title_string = soup.find('title').text
    title = title_string.split('(')[0].strip()
           
    rating = get_movie_value(soup,'MPAA Rating:')
        
    raw_release_date = get_movie_value(soup, 'Release Date')
    release_date = to_date(raw_release_date)

    raw_domestic_total_gross = get_movie_value(soup,'Domestic Total')
    domestic_total_gross = money_to_int(raw_domestic_total_gross)

    raw_runtime = get_movie_value(soup, 'Runtime')
    runtime = runtime_to_minutes(raw_runtime)
    
    raw_budget = get_movie_value(soup, 'Production Budget:')
    budget = convert_budget(raw_budget)
    
    the_subgenres = get_subgenre(title)
    
    director_string = soup.find('a', href= re.compile('Director&id'))
    try:
        director = director_string.text
    except AttributeError:
        director = None

    headers = ['title', 'domestic total gross',
           'release date', 'runtime (min)', 'rating', 'budget', 'subgenre', 'director']
        
    movie_dict = dict(zip(headers, [title,
                                    domestic_total_gross,
                                    release_date,
                                    runtime,
                                    rating,
                                    budget,
                                    the_subgenres,
                                    director
                                   ]))
    
    return(movie_dict)
```

### Add new feature: winning_director with info scraped from Wiki

```python
# Scrape Golden Globe winning directors from Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_Academy_Award_for_Best_Director_winners_by_age'
response = requests.get(url)
page = response.text

soup = BeautifulSoup(page, 'lxml')
td = soup.find_all('td')

tds = []
for x in td:
    tds.append(x.text)

nums = []
for t in range(1,94):
    nums.append(str(t))

winners = {}
i = 1
for item in tds:
    if item in nums:
        winners[tds[i].split(',')[0]] = tds[i + 3].split('(')[-1].split('-')[0]
    i += 1

winners
```

```python
# Create function to label a particular director as Golden Globe winner or not
def winning_director(row):
    try:
        last_name = row['director'].split(' ', 1)[1:][0] 
        if last_name in winners and row.year <= int(winners[last_name]):
            return(1)
        else:
            return(0)
    except IndexError:
        return(0)
```

```python
# Apply function to df
data['winning_director'] = data.apply(lambda row: winning_director(row), axis=1)
```

```python
# Drop duplicates
data.drop_duplicates(subset=['title', 'year'], inplace=True)
```

```python
# Write function to create 'season' feature based on 'month'
def which_season(row):
    winter = [12, 1, 2]
    spring = [3, 4, 5]
    summer = [6, 7, 8]
    fall = [9, 10, 11]
    if row['month'] in winter:
        return('winter')
    if row['month'] in spring:
        return('spring')
    if row['month'] in summer:
        return('summer')
    if row['month'] in fall:
        return('fall')
```

```python
# Create 'season' column
data['season'] = data.apply(lambda row: which_season(row), axis=1)
```

```python
# Convert categorical feature 'season' to dummy variables and add dummy variables to df
data = pd.concat([data.drop('season', axis=1), pd.get_dummies(data['season'])], axis=1)
```
